## Module 1: Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('songs_with_genres_complete.csv')
#fix accidental formatting issues
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 1)
df = df.dropna()
df.head()
```

```python
#what genres are most successful?
counts = df['genre'].value_counts()
othercount =sum(counts[10:])
counts = counts[:10]
counts['Other'] = othercount
# Plot pie chart
counts.plot(kind='pie', autopct='%1.1f%%', colors = ['#e21833', '#ffd200', '#454545'])
plt.title('Distribution of Genres')
```
